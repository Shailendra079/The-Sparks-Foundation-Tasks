we will make a subset of data to analyze it by each region

```python
data2 = data.groupby(["Region", "Category"]).sum()
data2 = data2.reset_index()
data2
```

Ploting the graph for each region v/s Quantity

```python
plt.figure(figsize=(12,4))
plt.title("Total Number of Products Sold in Each Region")
sb.set_style("whitegrid")
sb.barplot(x = 'Region', y =  'Quantity', hue= "Category", palette= 'rocket', data = data2);
```

Ploting the graph for each region v/s sales

```python
plt.figure(figsize=(12,4))
plt.title("Total Sales Made by Products in Each Region")

sb.set_style("whitegrid")
sb.barplot(x = 'Region', y =  'Sales', hue= "Category", data = data2);
```

```python
data2.hist(bins=30,figsize=(15,15))
plt.show()
```

 Now we will make another subset to analyze subcategory items by different parameters.


```python
data3 = data[["Subcategory", "Sales", "Profit","Quantity"]]
data3 = data3.groupby(["Subcategory"]).sum()
data3 = data3.reset_index()
data3 = data3.sort_values("Quantity", ascending= False)
data3
```

```python
plt.figure(figsize=(12,6))
plt.title("Number of items sold")
plt.xticks(rotation = 45)
sb.barplot(x= "Subcategory", y= "Quantity", data = data3);
```

```python
data4= data3[["Subcategory", "Sales", "Profit"]]
data4.set_index("Subcategory", inplace= True)
data4
```

Ploting the graph to find the relationship between sales and profit.

```python
plt.rcParams['figure.figsize']= (12,4)
data4.plot(kind='bar');
plt.title("Comparison between Sales and Profit of Products");
```

 Making another subset to analyze data by state

```python
data5 = data[["State", "Sales", "Profit", "Quantity", "Discount"]]
data5 = data5.groupby(["State"]).sum()
data5 = data5.reset_index()
data5 = data5.sort_values("Quantity", ascending= False)
data5.head(10)
```


Ploting the graph for top 10 state which sold maximum products.

```python
plt.rcParams['figure.figsize']= (12,4)
data5.head(10).plot(kind='bar', x= "State", y= "Quantity", color= 'green');
plt.title("Top 10 state which sold maximum products");
plt.ylabel("Number of products sold")
```

Ploting the graph for each state to know how much profit they made in comparison to sale.

```python
data6= data5[["State", "Sales", "Profit"]]
data6.set_index("State", inplace= True)

plt.rcParams['font.size']= 14
plt.rcParams['figure.figsize']= (12,4)
data6.head(10).plot(kind='bar');
plt.title("Comparison between Sales and Profit of Products in top 10 state");
```

Ploting the graph for maximum discount given by top 10 states.

```python
plt.rcParams['figure.figsize']= (12,4)
data5.head(10).sort_values("Discount", ascending = False).plot(kind='bar', x= "State", y= "Discount");
plt.title("Discount given by top 10 states");
plt.ylabel("Discount");
```
